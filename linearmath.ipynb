```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([6, 7, 7.5, 9, 10, 10.99, 12, 13.2, 14])
y = np.array([1, 3, 3, 5, 7, 7.5, 9, 11.4, 13.2])
```

```python
beta1_x = x.mean()
beta1_y = y.mean()

level_x = x - beta1_x
level_y = y - beta1_y
numerator = (level_x * level_y).sum()   
denominator = (level_x ** 2).sum()
beta1 = numerator / denominator
```

```python
#Beta0
beta0 = beta1_y - (beta1 * beta1_x)
y_pred = beta0 + beta1 * x
# result
print(f"Slope__Beta1  :   {beta1:.2f}")
print(f"y-intercept__Beta0  :   {beta0:.2f}")
```

```python
def show():
    plt.scatter(x, y, color='blue', label='Actual Data')
    plt.plot(x, y_pred, color='red', label='Linear Regression') 
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('simpel Linear Regression')
    plt.legend()
    plt.show()
show()
```
